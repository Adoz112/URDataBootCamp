```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Read the city data file
# File to Load (Remember to change these)
df = pd.read_csv("data/city_data.csv")
df.head()
```

```python
# Read the ride data file
df1 = pd.read_csv("data/ride_data.csv")
df1.head()
```

```python
# Combine the data into a single dataset
merge_table = pd.merge(df, df1, on="city")
merge_table
```

```python
# Display the data table for preview
merge_table = pd.merge(df, df1, on="city")
merge_table.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

ride_count_x = merge_table.groupby(['type', 'city'])['ride_id'].count()
ride_count_x
avg_fare_y = merge_table.groupby(['type', 'city'])['fare'].mean()
avg_fare_y



# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
```

```python
# Build the scatter plots for each city types
plt.scatter(ride_count_x['Rural'], avg_fare_y['Rural'], s=(ride_count_x['Rural']*50), marker="o", facecolors="yellow", edgecolors="black")
plt.scatter(ride_count_x['Urban'], avg_fare_y['Urban'], s=(ride_count_x['Urban']*50), marker="o", facecolors="red", edgecolors="black")
plt.scatter(ride_count_x['Suburban'], avg_fare_y['Suburban'], s=(ride_count_x['Suburban']*50), marker="o", facecolors="lightblue", edgecolors="blue")

#plt.scatter()


#avg_fare_y["Rural"]
```

```python
# Show plot
plt.show()
```
